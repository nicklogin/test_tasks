```python
from pymorphy2 import MorphAnalyzer
```

```python
from bs4 import BeautifulSoup
```

```python
from pymorphy2.tokenizers import simple_word_tokenize
```

```python
morph = MorphAnalyzer()
```

```python
with open('dom.txt','r',encoding='utf-8') as inp:
    text = inp.read()
```

а) лемматизировать текст с помощью mystem или pymorphy2

```python
all_tokens = simple_word_tokenize(text)
lemmas = [morph.parse(token)[0].normal_form for token in all_tokens]
lemmas[:30]
```

 б) найти такие леммы (не исходные словоформы), в которых было бы две (не больше, не меньше) буквы «о»

```python
lemmas=set(lemmas)
```

```python
double_o_lemmas = [lemma for lemma in lemmas if lemma.count("о")==2]
double_o_lemmas[:10]
```

```python
with open('double_o_lemmas.txt','w',encoding='utf-8') as outp:
    outp.write('\n'.join(double_o_lemmas))
```

в) программно обратиться к странице (скачать питоном) http://lib.ru/POEZIQ/PESSOA/lirika.txt 

```python
import urllib.request, json
```

```python
with urllib.request.urlopen("http://lib.ru/POEZIQ/PESSOA/lirika.txt") as inp:
    text1 = inp.read().decode('koi8-r')
```

```python
soup = BeautifulSoup(text1)
```

```python
clear_text1 = soup.get_text()
```

г) составить словарь для текста, который там размещен

```python
clear_text_lemmas = set([morph.parse(token)[0].normal_form for token in simple_word_tokenize(clear_text1) if morph.parse(token)[0].tag.POS])
clear_text_lemmas = list(sorted(clear_text_lemmas))
```

```python
clear_text_lemmas
```

д) записать этот словарь в файл в формате JSON

```python
with open('poet_words.json','w',encoding='utf-8') as outp:
    json.dump(clear_text_lemmas,outp,indent=2,ensure_ascii=False)
```
